```python
filt=df["Rating"] >= 4.5
df_yuksek=df[filt].sort_values(by="Install")

```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(df_yuksek["Install"], df_yuksek["Reviews"])
plt.style.use("fivethirtyeight")
plt.xlabel("Install")
plt.ylabel("Reviews")
plt.title("Install-Reviews Plot")

plt.show()
```

```python
plt.style.use("fivethirtyeight")

plt.title("Raiting - Install Mean")


df_yuksek=df[df["Rating"] >= 3.5]["Install"].mean()
df_orta=df[df["Rating"] < 3.5]["Install"].mean()
slices=[df_yuksek,df_orta]
labels = ["A","B"]
colors = ["blue", "red"]

plt.pie(slices, labels = labels, colors= colors, wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show( )
```

```python

plt.style.use("fivethirtyeight")

plt.title("Raiting-Reviews Mean")
df_yuksek=df[df["Rating"] >= 3.5]["Reviews"].mean()
df_orta=df[df["Rating"] < 3.5]["Reviews"].mean()
slices=[df_yuksek,df_orta]
labels = ["A","B"]
colors = ["blue", "red"]

plt.pie(slices, labels = labels, colors= colors, wedgeprops={"edgecolor":"black"})

plt.tight_layout()

plt.show( )
```
